```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(drop='first')
```

```python
enc.fit(X[["touch_screen"]])
```

```python
X_cat_ohe = pd.DataFrame(data=enc.transform(X[["touch_screen"]]).toarray(),  columns=enc.get_feature_names_out())

X_cat_ohe
```

```python
X_new = pd.concat([X.drop("touch_screen", axis=1), X_cat_ohe], axis=1)
X_new.columns
```

```python
test_cat_ohe = pd.DataFrame(data=enc.transform(df_test[["touch_screen"]]).toarray(),  columns=enc.get_feature_names_out())
test_data = pd.concat([df_test.drop("touch_screen", axis=1), test_cat_ohe], axis=1)
```

```python
test_data
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, f1_score
from sklearn.model_selection import cross_val_score

f1_weighted = make_scorer(f1_score, average='weighted')
```

```python
clf = LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1, class_weight=None, solver='lbfgs', max_iter=100, multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None, random_state=42)
```

```python
res = cross_val_score(clf, X_new, y, cv=3, scoring=f1_weighted)
```

```python
res.mean()
```

```python
from sklearn.model_selection import GridSearchCV

params = {'C': [10**n for n in range(-3, 3)]}
clf = LogisticRegression(random_state=42, penalty='l2', dual=False, tol=0.0001, fit_intercept=True, intercept_scaling=1, class_weight=None,  solver='lbfgs', max_iter=100, multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None)
gs = GridSearchCV(clf, params, scoring=f1_weighted, cv=3)
gs.fit(X_new, y)
```
